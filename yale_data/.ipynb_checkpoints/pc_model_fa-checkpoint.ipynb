```python
import numpy as np
import matplotlib.pyplot as plt
import copy
import glob
import imageio
import skimage
from skimage import data, io, filters
import pickle
import pandas as pd
import seaborn as sns

def get_labels(filenames):
    labels = np.zeros(len(filenames))
    for i in range(len(filenames)):
        if any(s in filenames[i] for s in ('sad', 'wink', 'surprised', 'sleepy', 'happy')):
            labels[i] = 1
    return labels

path = '/Users/madinaabdrakhmanova/Documents/GitHub/pc_fa/yale_data/'
img_arr_train = np.array([skimage.img_as_float(skimage.transform.rescale(imageio.imread(file),1.0 / 4.0)) for file in glob.glob(path+'yale_train/*png')])
img_arr_test = np.array([skimage.img_as_float(skimage.transform.rescale(imageio.imread(file),1.0 / 4.0)) for file in glob.glob(path+'yale_test/*png')])
X_train = np.reshape(img_arr_train, (img_arr_train.shape[0], img_arr_train.shape[1]*img_arr_train.shape[2]))
X_test = np.reshape(img_arr_test, (img_arr_test.shape[0], img_arr_test.shape[1]*img_arr_test.shape[2]))
y_train = get_labels(glob.glob(path+'yale_train/*png'))
y_test = get_labels(glob.glob(path+'yale_test/*png'))
data = (X_train, y_train, [])
test_data = (X_test, y_test, [])
```

```python
model = PCModel(data=data, model='fa', likelihood='default', predictor='logistic', backend='torch', lam = 1e4, reg_weight=1e2, reg_weight_fa=1e2)
```

```python
pc_config = '''
# Model options
model fa
lam 0 1 100 1000 10000 100000
reg_weight_fa LOGUNIFORM -5 5

# Prediction options
predictor logistic
reg_weight LOGUNIFORM -5 5
pred_bias 1

#Data options
cv 0

#Optimization options
backend torch 
'''
```

```python
model.initialize()
```

```python
pi, phi, eta = model.params
f = pi[0].flatten()
w = np.zeros(3)
w[1:] = eta[0]
init_params = np.concatenate((f, np.concatenate((pi[1][0].reshape((1)), np.concatenate((pi[2], w))))))
with open("init_params.pkl", 'wb') as f_save:
    pickle.dump([init_params], f_save)
```

```python
model.objective()
```

```python
model.fit()
```

```python
model.accuracy(test_data)
```
